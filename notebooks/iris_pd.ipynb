# Manipulating Iris Data with Pandas

```python
import io
import pandas as pd
import requests
```

```python
#importing the iris data from url
iris = requests.get('http://eaton-lab.org/data/iris-data-dirty.csv').content
iris = iris.decode()
```

```python
iris_pd = pd.read_csv(io.StringIO(iris), names = ("sep.l", "sep.w", "pet.l", "pet.w", "species"))
iris_pd.head()
```

```python
iris_clean = iris_pd.replace("Iris-setsa", "Iris-setosa").replace("Iris-versicolour", "Iris-versicolor").dropna()
iris_clean
```

```python
iris_clean.describe()
```

```python
for sp_var, sp_name in iris_clean.groupby("species"):
    print(f"{sp_var}")
    print(sp_name.describe())
```
